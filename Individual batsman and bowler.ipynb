```python
import pandas as pd
```

```python
df=pd.read_csv('FINAL.csv')
```

```python
df.columns
```

```python
df['isOne'] = df['runs_off_bat'].apply(lambda x: 1 if x == 1 else 0)
df['isTwo'] = df['runs_off_bat'].apply(lambda x: 1 if x == 2 else 0)
df['isThree'] = df['runs_off_bat'].apply(lambda x: 1 if x == 3 else 0)
df['isFour'] = df['runs_off_bat'].apply(lambda x: 1 if x == 4 else 0)
df['isSix'] = df['runs_off_bat'].apply(lambda x: 1 if x == 6 else 0)
df['Boundary'] = df['runs_off_bat'].apply(lambda x: 1 if x == 6 or x==4 else 0)
    
```

```python
df['wicket_type']=df['wicket_type'].fillna(0)
for index,row in enumerate(df['wicket_type']):
    if row=='obstructing the field' or row=='retired hurt' or row=='run out' or row==0:
        df.loc[index,'wicket']=0
    else:
        df.loc[index,'wicket']=1
```

```python
df=df[(df.innings==1)|(df.innings==2)]
```

```python
df.head()
```

```python
df2=df.groupby(['striker']).filter(lambda x: len(x['match_id'].unique())>20)
```

```python
df2['striker'].unique()
```

# Batsman against different bowling types and in different stages

```python
######  Enter the name of the batsman from above list   ##########

batter_name=input('Enter the batsmans name: ')
df1=df[df['striker']==batter_name]
df_bowlertype=df1.groupby(['bowler_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum','Boundary':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_faced','wicket':'dismissed'},axis=1)
df_bowlertype['Average']=df_bowlertype['Runs_scored']/df_bowlertype['dismissed']
df_bowlertype['Strike rate']=(df_bowlertype['Runs_scored']/df_bowlertype['Balls_faced'])*100
df_bowlertype['Dot ball percentage']=((df_bowlertype['number_of_dotballs']/df_bowlertype['Balls_faced'])*100).round()
df_bowlertype['Boundary Percentage']=((df_bowlertype['Boundary']/df_bowlertype['Balls_faced'])*100).round()
df_bowlertype=df_bowlertype.drop(['number_of_dotballs','Boundary'],axis=1)

display(df_bowlertype)

df_overtype=df1[df1['innings']==1]
df_overtype=df_overtype.groupby(['over_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum','isFour':'sum','isSix':'sum','Boundary':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_faced','wicket':'dismissed','isFour':"Fours",'isSix':'Sixes'},axis=1)
df_overtype['Average']=df_overtype['Runs_scored']/df_overtype['dismissed']
df_overtype['Strike rate']=(df_overtype['Runs_scored']/df_overtype['Balls_faced'])*100
df_overtype['Dot ball percentage']=((df_overtype['number_of_dotballs']/df_overtype['Balls_faced'])*100).round()
df_overtype['Boundary Percentage']=((df_overtype['Boundary']/df_overtype['Balls_faced'])*100).round()
df_overtype['Balls per boundary']=((df_overtype['Balls_faced']/df_overtype['Boundary']))

df_rpi=df1[df1['innings']==1]
df_rpi1=df_rpi['runs_off_bat'].sum()/df_rpi['match_id'].nunique()
#display(Runs per innings(first_innins),df_rpi1)
df_overtype1=df_overtype.drop(['number_of_dotballs','Boundary'],axis=1)
print('\t\t\t\t'+batter_name+' First Innings')
display(df_overtype1)


df_overtype=df1[df1['innings']==2]
df_overtype=df_overtype.groupby(['over_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum','isFour':'sum','isSix':'sum','Boundary':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_faced','wicket':'dismissed','isFour':"Fours",'isSix':'Sixes'},axis=1)
df_overtype['Average']=df_overtype['Runs_scored']/df_overtype['dismissed']
df_overtype['Strike rate']=(df_overtype['Runs_scored']/df_overtype['Balls_faced'])*100
df_overtype['Dot ball percentage']=((df_overtype['number_of_dotballs']/df_overtype['Balls_faced'])*100).round()
df_overtype['Boundary percentage']=((df_overtype['Boundary']/df_overtype['Balls_faced'])*100).round()
df_overtype['Balls per boundary']=((df_overtype['Balls_faced']/df_overtype['Boundary']))
df_overtype2=df_overtype.drop(['number_of_dotballs','Boundary'],axis=1)

df_rpi=df1[df1['innings']==2]
df_rpi2=df_rpi['runs_off_bat'].sum()/df_rpi['match_id'].nunique()

print('\t\t\t\t'+batter_name+' Second Innings')
display(df_overtype2)

print("\nRuns per innings(1st innings)",df_rpi1.round())
print("\nRuns per innings(2nd innings)",df_rpi2.round())
```

# Bowlers against different batting types and in different stages

```python
############# enter the bowler name from the above list

bowler_name=input('Enter the bowler name: ')
df2=df[df['bowler']==bowler_name]
df_bowler=df2.groupby(['Batting_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_bowled','wicket':'wickets'},axis=1)
df_bowler['bowling average']=df_bowler['Runs_scored']/df_bowler['wickets']
df_bowler['bowling strike-rate']=df_bowler['Balls_bowled']/df_bowler['wickets']
df_bowler['Dot ball percentage']=((df_bowler['number_of_dotballs']/df_bowler['Balls_bowled'])*100).round()

print('\n\t\t\t'+bowler_name+' against left hand and right hand batsman')
display(df_bowler)

df_overtype=df2[df2['innings']==1]
df_overtype=df_overtype.groupby(['over_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_bowled','wicket':'wickets'},axis=1)
df_overtype['bowling average']=df_overtype['Runs_scored']/df_overtype['wickets']
df_overtype['bowling strike-rate']=df_overtype['Balls_bowled']/df_overtype['wickets']
df_overtype['Dot ball percentage']=((df_overtype['number_of_dotballs']/df_overtype['Balls_bowled'])*100).round()
overs_bowled=[]
for i in df_overtype['Balls_bowled']:
    out1=i//6
    #print(str(out1))
    out2=i%6
    #print(out2)
    final=str(out1).split('.')[0]+'.'+str(out2).split('.')[0]
    overs_bowled.append(final)
df_overtype['overs_bowled']=overs_bowled
df_overtype['overs_bowled']=df_overtype['overs_bowled'].astype('float')
df_overtype['economy']=df_overtype['Runs_scored']/df_overtype['overs_bowled']

df_overtype1=df_overtype.drop(['Balls_bowled','number_of_dotballs'],axis=1)

print('\t\t\t\t'+bowler_name+' in First Innings')
display(df_overtype1)

df_overtype=df2[df2['innings']==2]
df_overtype=df_overtype.groupby(['over_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket':'sum','number_of_dotballs':'sum'}).reset_index().rename({'runs_off_bat':'Runs_scored','match_id':'Balls_bowled','wicket':'wickets'},axis=1)
df_overtype['bowling average']=df_overtype['Runs_scored']/df_overtype['wickets']
df_overtype['bowling strike-rate']=df_overtype['Balls_bowled']/df_overtype['wickets']
df_overtype['Dot ball percentage']=((df_overtype['number_of_dotballs']/df_overtype['Balls_bowled'])*100).round()
overs_bowled=[]
for i in df_overtype['Balls_bowled']:
    out1=i//6
    #print(str(out1))
    out2=i%6
    #print(out2)
    final=str(out1).split('.')[0]+'.'+str(out2).split('.')[0]
    overs_bowled.append(final)
df_overtype['overs_bowled']=overs_bowled
df_overtype['overs_bowled']=df_overtype['overs_bowled'].astype('float')
df_overtype['economy']=df_overtype['Runs_scored']/df_overtype['overs_bowled']

df_overtype2=df_overtype.drop(['Balls_bowled','number_of_dotballs'],axis=1)
print('\t\t\t\t'+bowler_name+' in Second Innings')
display(df_overtype2)

```
