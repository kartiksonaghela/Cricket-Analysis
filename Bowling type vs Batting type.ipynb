```python
import pandas as pd
```

```python
df=pd.read_csv('FINAL.csv')
```

```python
df['innings'].unique()
```

```python
df1=df.groupby(['bowler_type']).filter(lambda x: len(x['match_id'].unique())>15)

df1=df.groupby(['bowler_type','Batting_type']).agg({'runs_off_bat':'sum','match_id':'count','wicket_type':'count','number_of_dotballs':'sum'}).reset_index().rename({'match_id':'Total_balls','runs_off_bat':'Runs_scored','wicket_type':'wickets'},axis=1)
overs_bowled=[]
for i in df1['Total_balls']:
    out1=i//6
    #print(str(out1))
    out2=i%6
    #print(out2)
    final=str(out1).split('.')[0]+'.'+str(out2).split('.')[0]
    overs_bowled.append(final)
df1['overs_bowled']=overs_bowled
df1['overs_bowled']=df1['overs_bowled'].astype('float')
df1['economy']=df1['Runs_scored']/df1['overs_bowled']
df1['bowling average']=df1['Runs_scored']/df1['wickets']
df1['Dot ball percentage']=((df1['number_of_dotballs']/df1['Total_balls'])*100).round()

df1=df1.drop(['Total_balls','number_of_dotballs'],axis=1)
grouped_df = df1.groupby('bowler_type')    
gb = grouped_df.groups
df1=df1.drop(['bowler_type'],axis=1)
for key, values in gb.items():
    print('\t\t\t\t'+key+' Stats')
    #df1=df1.drop(['bowler_type'])
    display(df1.iloc[values])#, "\n\n")
```
